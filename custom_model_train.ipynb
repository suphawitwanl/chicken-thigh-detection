```python
import os
os.mkdir('extrated_here')
```

```python
import shutil

zip_file = '/content/drive/MyDrive/หยิบไก่/รูปน่องไก่/datasets_train/datasets.zip'
destination_directory = '/content/extrated_here'

shutil.unpack_archive(zip_file, destination_directory)
```

```python
import yaml

# Load the YAML data from the original file
with open('/content/extrated_here/datasets/data.yaml', 'r') as file:
    data = yaml.safe_load(file)

# Modify the paths in the 'train', 'val', and 'test' fields
data['train'] = data['train'].replace('../', '')
data['val'] = data['val'].replace('../', '')
data['test'] = data['test'].replace('../', '')


for key, value in data.items():
  print(f"{key} \t {value}")

# Save the modified data back to a new YAML file
with open('/content/extrated_here/datasets/data.yaml', 'w') as file:
    yaml.dump(data, file)
```

```python
from ultralytics import YOLO

# Load a model
model = YOLO('yolov8s.pt')  # load a pretrained model (recommended for training)

# Train the model
results = model.train(data='/content/extrated_here/datasets/data.yaml', epochs=200, imgsz=640)

#change timezone
import datetime
import pytz
dt = datetime.datetime.now()
tz = pytz.timezone('Asia/Bangkok')
new_dt = dt.astimezone(tz)
str_date = new_dt.strftime("%Y_%m_%d_%HH_%MM_%SS")
print(str_date)

#save to google drive
os.rename('/content/runs', f'/content/runs_{str_date}')
shutil.move(f'/content/runs_{str_date}', '/content/drive/MyDrive/หยิบไก่/runs_result')
```
