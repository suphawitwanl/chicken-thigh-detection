```python
from ultralytics import YOLO
import cv2
import os
from PIL import Image
from google.colab.patches import cv2_imshow
import numpy as np
```

```python

# Load a model
model = YOLO('/content/drive/MyDrive/หยิบไก่/รูปน่องไก่/V7/best_v7.pt')

print("1")
directory = '/content/drive/MyDrive/หยิบไก่/รูปน่องไก่/รูปไก่อันใหม่'
file_names = [file for file in os.listdir(directory) if os.path.isfile(os.path.join(directory, file))]
file_names.sort()
print(file_names)

print("2")
img_list = []
for file_n in file_names:
    img_list.append(f"/content/drive/MyDrive/หยิบไก่/รูปน่องไก่/รูปไก่อันใหม่/{file_n}")
print(img_list)

#print("4")
# Process results list
```

```python
#predict model

count = 143
print("3")
# Run batched inference on a list of images
for imgg in img_list:

  image = cv2.imread(imgg, 0)
  gausBlur = cv2.GaussianBlur(image, (35,35),0)
  rgb_image = np.stack((gausBlur,)*3, axis=-1)


  results = model(rgb_image, conf=0.7)  # return a list of Results objects


  for r in results:
    count += 1
    im_array = r.plot()  # plot a BGR numpy array of predictions
    im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
    #im.show()  # show image
    im.save(f'/content/drive/MyDrive/หยิบไก่/รูปน่องไก่/V7/result_v7/results{count}.jpg')  # save image



print("5 end")
```
